### Importing needed libraries and loading the data

```python
import joblib
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#Importing x data
X_train = pd.read_csv('data/x_train.csv', header=None).drop([0], axis=1)
X_train.drop([0], axis=0, inplace = True)
X_train = X_train.values

x_test = pd.read_csv('data/x_test.csv', header=None).drop([0], axis=1)
x_test.drop([0], axis=0, inplace = True)
x_test = x_test.values

#Importing y data
y_train = pd.read_csv('data/y_train.csv', header=None, usecols = [1], skiprows=1)
y_train = y_train.values

y_test = pd.read_csv('data/y_test.csv', header=None,  usecols = [1], skiprows=1)
y_test = y_test.values
```

### Training hyperparameters of the decision tree model using GridSearchCV

```python
param_grid = [{'max_depth':range(1, 15),
               'max_features':range(1, 8),
               'min_samples_split':range(2, 5),
               'min_samples_leaf' : range(1, 5), 
               'splitter': ['best'], 
               'criterion': ['gini', 'entropy']}]

tree_clf = DecisionTreeClassifier(random_state=1)
grid_search = GridSearchCV(tree_clf, param_grid, cv=5, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
print(f'Best parameters are: {grid_search.best_params_} \n')
print(f'Best estimator is: {grid_search.best_estimator_} \n')
print(f'Best score is {grid_search.best_score_}')
```

### Training the model with parameters obtained from GridSearchCV and predicting on test set

```python
tree_clf = DecisionTreeClassifier(max_depth=7, 
                                  max_features=6,
                                  min_samples_split=4, 
                                  min_samples_leaf=1, 
                                  splitter='best',
                                  criterion='gini', 
                                  random_state=1) 
tree_clf.fit(X_train, y_train)
y_pred = tree_clf.predict(x_test)
```

### Evaluation of model performance - Accuracy and Confusion Matrix

```python
dt_accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy score on test set is {dt_accuracy}')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
dt_cmatrix= confusion_matrix(y_test, y_pred)
sns.heatmap(dt_cmatrix.T, square=True, annot=True, fmt='d', cbar=False, ax=ax)
ax.yaxis.set_ticklabels(['N.America','S.America','Europe','Africa','Asia','Oceania'])
ax.xaxis.set_ticklabels(['N.America','S.America','Europe','Africa','Asia','Oceania'])
plt.title('Confusion matrix of Decision Tree Algorithm')
plt.xlabel('True label')
plt.ylabel('Predicted label')
```
